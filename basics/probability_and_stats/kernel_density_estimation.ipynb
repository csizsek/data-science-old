### Generate the Gaussian Mixture that is be Estimated

```python
m1 = 3.0
m2 = -2.0
m3 = 0.0

s1 = 0.5
s2 = 0.2
s3 = 1.2

n1 = 100
n2 = 100
n3 = 300
```

```python
def gaussian(x, mu, sig2):
    return (1.0 / np.power(2 * sig2 * np.pi, 0.5)) * np.exp(-(np.power(x - mu, 2.)) / (2.0 * sig2))
```

```python
ls = np.linspace(-4, 5, 200)
g1 = gaussian(ls, m1, s1) / 5.0
g2 = gaussian(ls, m2, s2) / 5.0
g3 = gaussian(ls, m3, s3) * 3.0 / 5.0
g = g1 + g2 + g3
```

```python
_ = plt.figure(figsize=fsize)
_ = plt.plot(
    ls,
    g,
    c='blue',
    linewidth=linewidth,
    label='Gaussian Mixture')

_ = plt.xlim((-4, 5))
_ = plt.ylim((-0.05, 0.5))
_ = plt.title('Gaussian Mixture')
_ = plt.xlabel('x')
_ = plt.ylabel('p(x)')
_ = plt.legend(loc='upper right')
```

### Generate a Sample from the Gaussian Mixture

```python
sample = np.concatenate([
        np.random.randn(n1) * s1 + m1,
        np.random.randn(n2) * s2 + m2,
        np.random.randn(n3) * s3 + m3])
```

```python
_ = plt.figure(figsize=fsize)
_ = plt.plot(
    ls,
    g,
    c='blue',
    linewidth=linewidth,
    label='Gaussian Mixture')

_ = plt.scatter(
    x=sample,
    y=[0]*len(sample),
    marker='|',
    c='red',
    s=100,
    label='Sample',
    linewidth=linewidth)

_ = plt.xlim((-4, 5))
_ = plt.ylim((-0.05, 0.5))
_ = plt.title('Gaussian Mixture Sample')
_ = plt.xlabel('x')
_ = plt.ylabel('p(x)')
_ = plt.legend(loc='upper right')
```

### Gaussian KDE with Different Bandwith Parameters

```python
bandwidths = [0.05, 0.4, 1.5]
kde_labels = [
    'KDE (Too Noisy)',
    'KDE (Good Fit)',
    'KDE (Too Smooth)']
kde_colors = [
    'orange',
    'green',
    'purple']
kdes = []

for i in range(3):
    kdes.append(np.exp(
        sklearn.neighbors.KernelDensity(
            bandwidth=bandwidths[i],
            kernel='gaussian')
        .fit(sample[:, np.newaxis])
        .score_samples(ls[:, np.newaxis])))
```

```python
_ = plt.figure(figsize=fsize)
_ = plt.plot(
    ls,
    g,
    c='blue',
    linewidth=linewidth,
    label='Gaussian Mixture (Original)')

_ = plt.scatter(
    x=sample,
    y=[0]*len(sample),
    marker='|',
    c='red',
    s=100,
    label='Sample',
    linewidth=linewidth)

for i in range(3):
    _ = plt.plot(
        ls,
        kdes[i],
        c=kde_colors[i],
        linewidth=linewidth,
        label=kde_labels[i])

_ = plt.xlim((-4, 5))
_ = plt.ylim((-0.05, 0.5))
_ = plt.title('Gaussian KDE')
_ = plt.xlabel('x')
_ = plt.ylabel('p(x)')
_ = plt.legend(loc='upper right')
```
