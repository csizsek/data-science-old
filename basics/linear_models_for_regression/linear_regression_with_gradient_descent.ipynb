### Define the True Function and Generate the Sample Data

```python
alpha_true = 0.5
beta_true = 1
n = 50
x = np.random.uniform(0, 5.5, n)
y_true = alpha_true * x + beta_true
t = alpha_true * x + beta_true + (np.random.rand(n) - 0.5)
```

```python
_ = plt.figure(figsize=fsize)
_ = plt.scatter(x, t, color='b', marker='o', s=dotsize, label='Sample Data')
_ = plt.plot([np.min(x), np.max(x)], [np.min(y_true), np.max(y_true)], color='g', linewidth=linewidth, label='True Function')
_ = plt.xlim((0, 5.5))
_ = plt.ylim((0, 5.5))
_ = plt.xlabel('x')
_ = plt.ylabel('y')
_ = plt.title('True Function')
_ = plt.legend(loc='upper right')
```

### Implement Gradient Descent

```python
def error(a, b):
    s = 0
    for i in range(n):
        s += math.pow((a*x[i] + b) - t[i], 2)
    return 0.5 * s
```

```python
@np.vectorize
def errorv(a, b):
    return error(a, b)
```

```python
ran_alpha = np.linspace(-3, 5, 100)
ran_beta = np.linspace(-5, 10, 100)
ran_alpha, ran_beta = np.meshgrid(ran_alpha, ran_beta)
e = errorv(ran_alpha, ran_beta)
```

```python
ax = plt.figure(figsize=fsize).add_subplot(111, projection='3d')
_ = ax.plot_wireframe(ran_alpha, ran_beta, e, linewidth=linewidth, label='Error Function')
_ = ax.set_xlabel('alpha')
_ = ax.set_ylabel('beta')
_ = ax.set_zlabel('E(alpha, beta)')
_ = plt.title('Error Function Surface')
_ = plt.legend(loc='upper right')
```

```python
levels = [10, 30, 100, 300, 1000, 3000, 10000]
_ = plt.figure(figsize=fsize)
_ = plt.contourf(ran_alpha, ran_beta, e, locator=mpl.ticker.LogLocator(2), label='Error Function Contour')
_ = plt.xlabel('alpha')
_ = plt.ylabel('beta')
_ = plt.xlim((-3, 5))
_ = plt.ylim((-5, 10))
_ = plt.title('Error Function Contours')
_ = plt.legend(loc='upper right')
```

```python
def dda_error(a, b):
    s = 0
    for i in range(n):
        s += x[i] * (a*x[i] + b - t[i])
    return s
```

```python
def ddb_error(a, b):
    s = 0
    for i in range(n):
        s += a*x[i] + b - t[i]
    return s
```

### Perform Gradient Descent

```python
error_threshold = 0.0000001
step_size = 0.0001

alpha_old = 0
beta_old = 0

alpha_new = 0.1
beta_new = 0.1

i = 0

while abs(error(alpha_old, beta_old) - error(alpha_new, beta_new)) > error_threshold:
    i += 1
    
    alpha_old = alpha_new
    beta_old = beta_new
    
    alpha_new = alpha_old - step_size * dda_error(alpha_old, beta_old)
    beta_new = beta_old - step_size * ddb_error(alpha_old, beta_old)
```

```python
print('iterations:           ', i)
print('alpha approximation:  ', alpha_new)
print('beta approximation:   ', beta_new)
```

```python
y_new = alpha_new * x + beta_new
_ = plt.figure(figsize=fsize)
_ = plt.scatter(x, t, color='b', marker='o', s=dotsize, label='Sample Data')
_ = plt.plot([np.min(x), np.max(x)], [np.min(y_true), np.max(y_true)], color='g', linewidth=linewidth, label='True Function')
_ = plt.plot([np.min(x), np.max(x)], [np.min(y_new), np.max(y_new)], color='r', linewidth=linewidth, label='Estimated Relationship')
_ = plt.xlim((0, 5.5))
_ = plt.ylim((0, 5.5))
_ = plt.xlabel('x')
_ = plt.ylabel('y')
_ = plt.title('True and Estimated Functions')
_ = plt.legend(loc='upper right')
```
